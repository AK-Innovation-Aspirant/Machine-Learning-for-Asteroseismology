```python
g_max = np.max(g_mode_spacing)
g_min = np.min(g_mode_spacing)
g_min
```

```python
g_max
```

```python
g_range = np.linspace(g_min,g_max,50)
g_range
```

```python
g_range_bins = np.empty([198004,50])
for i in range(0,198004):
    g_modes = g_mode_spacing
    initial_array = np.zeros(g_range.shape)
    diff = np.abs(g_range-g_modes[i])
    initial_array[np.argmin(diff)] = 1
    g_range_bins[i] = initial_array
g_range_bins
```
