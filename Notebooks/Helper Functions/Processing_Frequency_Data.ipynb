```python
#ran test with 5 samples hence the output.
freq_max = np.array([])
freq_min = np.array([])
for i in range(0,198004):
    max_val = np.max(frequency_data[i])
    freq_max = np.append(freq_max,max_val)
    min_val = np.min(frequency_data[i])
    freq_min = np.append(freq_min,min_val)
freq_max
```

```python
freq_min
```

```python
freq_max_val = np.max(freq_max)
freq_min_val = np.min(freq_min)

```

```python
star_range = np.linspace(freq_min_val,freq_max_val,30000)
star_range
```

```python
import matplotlib.pyplot as plt

#Example of two eigen frequencies
freqs = np.array([1.492,2.013])

temp_range = np.linspace(0.5,5,30000)
a  = np.zeros(temp_range.shape)
for i in range(freqs.shape[0]):
    dist = np.abs(temp_range - freqs[i])
    a[np.argmin(dist)]=1
plt.plot(temp_range,a)
```

```python
labelled_points = np.empty([198004,30000])
for i in range(0,198004):
    freqs = frequency_data[i]
    initial_array = np.zeros(star_range.shape)
    for j in range(freqs.shape[0]):
        diff = np.abs(star_range-freqs[j])
        initial_array[np.argmin(diff)] = 1
    labelled_points[i] = initial_array
labelled_points
```
